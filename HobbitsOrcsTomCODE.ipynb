```python
class Cross_River:
    
    def __init__(self, hobbit_amnt, orc_amnt, tom_bab, boat_size = 0): #Sets attributes of state one 
        self.hobbit_amnt = hobbit_amnt
        self.orc_amnt = orc_amnt
        self.tom_bab = 1
        self.boat_size = boat_size
        # boat size will be set during init method - self.boat_size = boat_size
        # other_node is not needed because parameter will transfer // self.other_node = 0
    
    def get_boat_size(self):
        if int(self.hobbit_amnt) <= 3:
            print("Boat size is 2")
            self.boat_size = 2
            return self.boat_size

        if int(self.hobbit_amnt) == 4:
            print("Boat size is 3")
            self.boat_size = 3
            return self.boat_size

        if int(self.hobbit_amnt) == 5:
            print("Boat size is 3")
            self.boat_size = 3
            return self.boat_size

        if int(self.hobbit_amnt) > 5:
            print("Boat size is 4")
            self.boat_size = 4
            return self.boat_size
        
    def get_orc_amnt(self):
        return self.orc_amnt
        
    def transfer(self, other_node): # Might have to check both left and right side 
        # other_node will get assigned value when transfer () is called 
        # other_node should be an instance because thinking of nodes as left and right side of river
        
        if self.boat_size == 2:
            print("Hobbits on left side: ")
            print(self.hobbit_amnt)

            print("Orcs on left side: ")
            print(self.get_orc_amnt())

            if self.hobbit_amnt > self.orc_amnt: #Simulating the hobbits and orcs going over the river
                print("\nSince hobbit amount exceeds orc amount, Tom will take one hobbit with him.")
                self.hobbit_amnt -= 1
                other_node.hobbit_amnt += 1
            else:
                print("Tom Bombadil will row. Since orc amount exceeds or is equal to hobbit amount, Tom will take one orc with him")
                self.orc_amnt -= 1
                other_node.orc_amnt += 1

        elif self.boat_size == 3:
            #Simulating the hobbits and orcs going over the river
            if self.hobbit_amnt > self.orc_amnt:
                if self.hobbit_amnt > 0 and self.orc_amnt > 0:
                    print("Tom Bombadil will row. Since hobbit amount exceeds orc amount, Tom will take one hobbit and one orc with him.") 
                    self.hobbit_amnt -= 1
                    other_node.hobbit_amnt += 1
                    self.orc_amnt -= 1
                    other_node.orc_amnt += 1 
                elif self.hobbit_amnt > 0: 
                    print("There are no more orcs. Only take a hobbit.")
                    self.hobbit_amnt -= 1
                    other_node.hobbit_amnt += 1
                elif self.orc_amnt > 0:
                    print("There are no more hobbits. Only take a orc.")
                    self.orc_amnt -= 1
                    other_node.orc_amnt += 1 
                print("")
            else:
                print("Tom Bombadil will row. Since orc amount exceeds or is equal to hobbit amount, Tom will only take one hobbit with him.")
                self.hobbit_amnt -= 1
                other_node.hobbit_amnt += 1

        elif self.boat_size == 4: 
            if self.hobbit_amnt > self.orc_amnt:
                print("Tom Bombadil will row. Since hobbit amount exceeds orc amount, Tom will take two hobbits and one orc with him in the boat.")
                self.hobbit_amnt -= 2
                self.orc_amnt -= 1
                other_node.hobbit_amnt += 2
                other_node.orc_amnt += 1 
            elif self.hobbit_amnt == self.orc_amnt: 
                self.hobbit_amnt -= 2
                self.orc_amnt -= 1
                other_node.hobbit_amnt += 2
                other_node.orc_amnt += 1
                if other_node.hobbit_amnt < other_node.orc_amnt:
                    print("Only take an orc")
                    self.hobbit_amnt -= 1
                    self.orc_amnt -= 2
                    other_node.hobbit_amnt += 1
                    other_node.orc_amnt += 2
            elif self.hobbit_amnt > 0: 
                    print("Only take a hobbit.")
                    self.hobbit_amnt -= 1
                    other_node.hobbit_amnt += 1
            elif self.orc_amnt > 0:
                print("Only take an orc.")
                self.orc_amnt -= 1
                other_node.orc_amnt += 1 
        else:
                print("Tom Bombadil will row. Since orc amount exceeds or is equal to hobbit amount, Tom will only take one hobbit with him.")
                self.hobbit_amnt -= 1
                other_node.hobbit_amnt += 1
            
    
    # @property # not passing parameters - still runs code to return a value - purely based on attributes of class itself
    def empty(self):
        if self.hobbit_amnt == 0:
            if self.orc_amnt == 0:
                return True 
        return False
```

```python
def main():
    
    print("\n")
    print("\tStart Game: \nThe task is to move all from left to right side of the river.")
    print("\tRules:\nThe constraints within this problem outline there is always one less orc than hobbit.")
    print("The constraints within this problem outline there is only one Tom Bombadil no matter what. \nThe most efficient way to solve this problem is: ")
    
    #Logic
    print("Have Tom Bombadil row the boat back and forth until all the hobbits and orcs are across the river. (Tom will get a good workout in!)")
        
    hobbit_amnt = int(input("How many hobbits are there? (Please enter a positive integer (ex: 2)): "))
    orc_amnt = (hobbit_amnt) - 1
    tom_bab = 1
    
    print("\nStarting state:")
    for x in range(hobbit_amnt):
      print("H                    | --- | ")

    for x in range(orc_amnt):
      print("O                    | --- | ")
    
    print("Goal state:")
    for x in range(hobbit_amnt):
      print("  | --- |        H")

    for x in range(orc_amnt):
      print("  | --- |        O")
    
    print("Left side -> right side river travel. \nThe amount of hobbits and orcs on the left side is what is shown in the solution when 'main()' is run. \nFirst number is hobbit amount. Second number is orc amount.")
    print(" ")
    print("Solution: ")
    
    starting_state_left = Cross_River(hobbit_amnt, orc_amnt, tom_bab)
    starting_state_right = Cross_River(0,0,0)
    
    starting_state_left.get_boat_size() # on instance where created 
    
    while not starting_state_left.empty():
        print(starting_state_left.hobbit_amnt)
        print(starting_state_left.orc_amnt)
        starting_state_left.transfer(starting_state_right)
    # no "()" for property 
        
        #Call the main function
if __name__ == '__main__':
    main()
```
